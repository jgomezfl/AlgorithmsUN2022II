<h2> Recursive fibonacci

```python
import itertools as tools
import matplotlib.pyplot as plt
from time import time
```

```python
#Recursive function

def fibRe(n):
  if n==0 :
      return 0
  if n==1 :
      return 1
  return fibRe(n-1)+fibRe(n-2)

#Iterative function

def fibIt(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b

#Binary exponentiation

def fib_fast_matrix(n):

    if n==0: return 0

    # Initialize a matrix [[1,1],[1,0]]    
    v1, v2, v3 = 1, 1, 0  
    # Perform fast exponentiation of the matrix (quickly raise it to the nth power)
    for rec in bin(n)[3:]:
        calc = (v2*v2) 
        v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3) 
        if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
    return v2;    

    
```

```python


#Recursive Fibonacci time
RecursiveT = []
RecursiveN = []
for i in range(50):
  RecursiveN.append(i)
  start = time()
  fibRe(i)
  finish = time()
  exectime = finish-start
  RecursiveT.append(exectime)
plt.title("Recursive execution")
plt.plot(RecursiveN,RecursiveT,label="Fibonacci Recursivo")
plt.plot(RecursiveN,RecursiveN,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```

```python
plt.title("Ejecución recursiva")
plt.plot(RecursiveN,RecursiveT,label="Fibonacci Recursivo")
plt.plot(RecursiveN,RecursiveN,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```

```python
#Iterative execution
IterativeT = []
IterativeN = []
for i in range(1000):
  IterativeN.append(i)
  start = time()
  fibIt(i)
  finish = time()
  exectime = finish-start
  IterativeT.append(exectime)
plt.title("Fibonacci Iterativo")
plt.plot(IterativeN,IterativeT,label="Fibonacci Iterativo")
plt.plot(IterativeN,IterativeN,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```

```python

#Iterative logarithmic execution
IterativelogT = []
IterativelogN = []
for i in range(1000):
  IterativelogN.append(i)
  start = time()
  fib_fast_matrix(i)
  finish = time()
  exectime = finish-start
  IterativelogT.append(exectime)
plt.title("Fibonacci iterativo Log")
plt.plot(IterativelogN,IterativelogT,label="Fibonacci Iterativo logaritmico")
plt.plot(IterativelogN,IterativelogN,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```
